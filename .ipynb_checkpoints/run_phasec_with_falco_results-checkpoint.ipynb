```python
import numpy as np
import pickle
from astropy.io import fits
import astropy.units as u
from pathlib import Path
from IPython.display import clear_output
import falco
import proper
import roman_phasec_proper as phasec

import misc

falco_dir = Path('/groups/douglase/kians-data-files/falco-phasec-data/hlc-band1-dms')

from roman.EXAMPLE_config_Roman_CGI_HLC_NFOV_Band1 import mp
```

```python
with open('Series0001_Trial0001_HLC_snippet.pkl', 'rb') as f: data = pickle.load(f)
```

```python
dm1_ref = data.dm1.Vall[:,:,0]*mp.dm1.VtoH
dm2_ref = data.dm1.Vall[:,:,0]*mp.dm2.VtoH
misc.myimshow2(dm1_ref, dm2_ref)
```

```python
falco_dm1 = data.dm1.Vall[:,:,5]*mp.dm1.VtoH
falco_dm2 = data.dm2.Vall[:,:,5]*mp.dm2.VtoH

misc.myimshow2(falco_dm1, falco_dm2)
```

```python
cor_type='hlc'
lambda_m=575e-9
final_sampling_lam0=0.1
output_dim=256
source_x_offset=0
source_y_offset=0
use_fpm=1
use_errors=1
if use_errors==0: use_lens_errors=0
else: use_lens_errors=1
use_hlc_dm_patterns=0
polaxis=10
use_field_stop=1
use_pupil_defocus=0

# dm1_m = np.zeros((48,48))
# dm2_m = np.zeros((48,48))

# dm1_fname='hlc_best_contrast_dm1.fits'
# dm2_fname='hlc_best_contrast_dm2.fits'
# dm1_m = proper.prop_fits_read( phasec.lib_dir + '/examples/' + dm1_fname )
# dm2_m = proper.prop_fits_read( phasec.lib_dir + '/examples/' + dm2_fname )

dm1_m = 2*falco_dm1
dm2_m = 2*falco_dm2

lam_array = np.array([lambda_m*1e6])
(proper_psfs, psfs_sampling_m) = proper.prop_run_multi('roman_phasec', lam_array, output_dim, QUIET=False, 
                                                PASSVALUE={'cor_type':cor_type,
                                                           'final_sampling_lam0':final_sampling_lam0, 
                                                           'source_x_offset':source_x_offset,
                                                           'use_fpm':use_fpm,
                                                           'use_hlc_dm_patterns':use_hlc_dm_patterns,
                                                           'use_errors': use_errors,
                                                           'use_lens_errors':use_lens_errors,
                                                           'use_pupil_defocus':use_pupil_defocus,
                                                           'use_dm1':1, 'dm1_m':dm1_m, 
                                                           'use_dm2':1, 'dm2_m':dm2_m,
                                                           'use_field_stop':use_field_stop,
                                                           'polaxis':polaxis,
                                                          })
clear_output()

proper_psf = proper_psfs[0]
psf_pixelscale = psfs_sampling_m[0]*u.m/u.pix

misc.myimshow2(np.abs(proper_psf)**2, np.angle(proper_psf), lognorm1=True, pxscl1=psf_pixelscale, pxscl2=psf_pixelscale,)
```
